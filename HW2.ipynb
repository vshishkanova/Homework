```python
#Задание 1
#записываем стих в переменную
text = 'варкалось хливкие шорьки пырялись по наве и хрюкотали зелюки как мюмзики в мове о бойся бармаглота сын он так свирлеп и дик а в глyще рымит исполин злопастный брандашмыг'

text1 = text.split() #разбиваем на слова текст

final= [] #создаем список для биграмм
for i in range(len(text1)-1): #создаем цикл для составления биграмм для всех слов, за исключением последнего
    final.append([text1[i],text1[i+1]]) #создаем биграмму из данного слова плюс последующего и добавляем в список
print(final)
```
